```python
import yfinance as yf
import pandas as pd
import datetime

msft = yf.Ticker("MSFT")

stock_info = msft.info

pd.set_option('display.max_rows', None)

stock_df = pd.DataFrame.from_dict(stock_info, orient="index")

display(stock_df)
```

```python
stock_sector = stock_df[0]["sector"]

print(stock_sector)
```

```python
news = msft.news
```

```python
news_df = pd.DataFrame.from_dict(news)
news_df["providerPublishTime"] = pd.to_datetime(news_df["providerPublishTime"], unit = "s")
news_df.set_index("providerPublishTime", inplace=True)
```
